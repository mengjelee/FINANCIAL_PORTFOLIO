```python

#hw1-本 金 平 均 攤 還 法 試 算 利 息

p = float(input("本金 (萬元)")) #principal
y = float(input("期數(年)")) #period 期數(年)可為非負小數
if y <= 0:
    print("期數不可為零或是負數")
else:
    rate_input = float(input("年利率(%)")) #interest rate 可為任意值
    rate = rate_input/1200
    #-------------------------計算每月攤還本金
    p_pay = 0
    p_pay = round(float(p/(y*12)),3)  
    r_pay_list = []
    p_final_pay = round(p - p_pay*((y-1)*12+11),3)
    p_pay_list = []
    for q in range(int(y*12)-1):
        p_pay_list.append(p_pay)
    p_pay_list.append(p_final_pay)
    #---------------------------計算每月利息
    for i in range(int(y*12)):  
        r_pay = 0
        r_pay = round((p - p_pay*i)*rate,3)
        r_pay_list.append(r_pay)
    #---------------------------計算本金利息累計
    p_r_total = 0
    p_r_total_list = []
    for u in range(int(y*12)):
        p_r_total_list.append(p_pay_list[u] + r_pay_list[u])
    present_l = []

    for j in range(int(y*12)):
        present_l.append([str(j+1),p_pay_list[j],r_pay_list[j],round(sum(p_r_total_list[:j+1]),3)])

    #---------------------------------以下用matplotlib製作表格輸出
    import matplotlib.pyplot as plt
    import matplotlib
    import pandas as pd
    plt.figure(figsize=(12, 9))
    font = {'family' : 'DFKai-SB',
    'weight' : 'bold',
    'size'  : '20'} #字型設定
    plt.rc('font', **font) #字型設定 # pass in the font dict as kwargs
    plt.rc('axes',unicode_minus=False) #字型設定
    col_labels = ['期數','本金','利息','本金利息累計']
    row_colors = ['white','gray','silver','gold']
    plt.axis('off')
    present_l2 = pd.DataFrame(present_l, columns=col_labels)
    plt.table(cellText=present_l2.values, colLabels=present_l2.columns, colColours=row_colors,cellLoc = 'center', bbox=[1,1,1, y*1.2])
    plt.show() #以表格輸出
```
